# Translation (shifting of images)

```python
image = cv2.imread('C:\\Users\\admin\\Pictures\\search_analytics_image.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
import numpy as np

M = np.float32([[1, 0, 250], [0, 1, 0]]) # [x],[y]
# x is set to 1, y to zero in first bracket, with 3rd option we can do shifting. which is
# number of pixcel shifted.
# y is set to 1, x to zero in second bracket, with 3rd option we can do shifting.
# note: with negative value we can shift image to opposite direction
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(shifted)
```

# Masking

```python
import cv2
image = cv2.imread('C:\\Users\\admin\\Pictures\\search_analytics_image.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
plt.imshow(image)
```

```python
import numpy as np
mask = np.zeros(image.shape[:2], dtype = "uint8")
(cX, cY) = (image.shape[1] // 2, image.shape[0] // 2)
cv2.rectangle(mask, (cX - 75, cY - 75), (cX + 75 , cY + 75), 255,-1)
plt.imshow(mask,cmap='gray')
masked = cv2.bitwise_and(image, image, mask = mask)
plt.imshow(masked)

```

```python
mask = np.zeros(image.shape[:2], dtype = "uint8")
cv2.circle(mask, (cX, cY), 100, 255, -1)
masked = cv2.bitwise_and(image, image, mask = mask)
plt.imshow(mask)
plt.imshow(masked)
```
